```python
with open('AAPL_historical.csv') as csv_file:
    csvfile = csv.reader(csv_file, delimiter=",")
    all_value = []
    for row in csvfile:
        for i in range(len(row)):
            if row[i] == '':
                row[i] = None
            value = (row[0], row[1], row[2], row[3], row[4], row[5], row[6])
        all_value.append(value)

del all_value[0]
```
